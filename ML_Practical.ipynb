```
import pandas as pd
df = pd.read_csv('car_evaluation.csv')
```

```
df
```

```
dic = {'vhigh':'BuyingPrice',
       'vhigh.1':'MaintenanceCost',
       '2':'NOD',
       '2.1':'NOP',
       'small':'Boot',
       'low':'Safety',
       'unacc':'Decision'}
df.rename(columns = dic, inplace = True)
```

```
df.head()
```

```
from sklearn.preprocessing import LabelEncoder
```

```
BuyingPrice=LabelEncoder()
MaintenanceCost=LabelEncoder()
Boot=LabelEncoder()
Safety=LabelEncoder()
Decision=LabelEncoder()
```

```
df['BuyingPrice']=BuyingPrice.fit_transform(df['BuyingPrice'])
df['MaintenanceCost']=BuyingPrice.fit_transform(df['MaintenanceCost'])
df['Boot']=BuyingPrice.fit_transform(df['Boot'])
df['Safety']=BuyingPrice.fit_transform(df['Safety'])
df['Decision']=BuyingPrice.fit_transform(df['Decision'])
```

```
df
```

```
df['Decision'].unique()
```

```
df['NOP'].unique()
```

```
df['NOD'].unique()
```

```
df['Decision'].unique()
```

```
df['NOP'].replace('more','6',inplace=True)
```

```
df['NOD'].replace('5more','5',inplace=True)
```

```
featCol = ['BuyingPrice','MaintenanceCost','NOD','NOP','Boot','Safety']
X = df[featCol]
```

```
y = df.Decision
```

```
from sklearn.model_selection import train_test_split
```

```
Xtr,Xte,ytr,yte = train_test_split(X,y,test_size=0.2)
```

```
from sklearn.tree import DecisionTreeClassifier
```

```
# classifier = DecisionTreeClassifier(criterion='gini')
classifier = DecisionTreeClassifier(criterion='entropy')
```

```
classifier.fit(Xtr,ytr)
```

```
# BuyingPrice -> 0,1,2,3
# MaintenanceCost -> 0,1,2,3
# NOD	-> 2,3,4,5
# NOP	-> 2,4,6
# Boot ->	0,1,2
# Safety ->	0,1,2
# Decision -> 0,1,2,3
```

```
classifier.predict([[3,3,2,2,0,0]])
```

```
classifier.predict(Xte)
```

```
classifier.score(Xte,yte)
```

```
from sklearn import tree
tree.plot_tree(classifier)
```
